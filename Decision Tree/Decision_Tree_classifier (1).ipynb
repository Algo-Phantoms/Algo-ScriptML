### Importing Basic Libraries

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Importing Dataset

````
dataset = pd.read_csv("diabetes.csv")
````

### Data Preparation 

````
dataset.head()
````

````
dataset.shape
````

````
dataset.isna().sum()
````

### Feature Selection

#### 1) Visualizing our dataset features

````
corr = dataset.corr()
sns.heatmap(dataset, cmap = "YlGnBu")
````

#### 2) Splitting the dataset into independent and dependent variables

````
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
````

````
x
````

````
y
````

### Splitting Dataset into trainig and testing set

````
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.25,random_state = 0) 
````

### Building Decision Tree Model

#### 1) Training Model

````
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="entropy",max_depth=3)
clf = model.fit(x_train,y_train)
````

#### 2) Testing Predicting Results

````
y_pred = model.predict(x_test)
````

### Evaluating Model Performance

````
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Confusion  Matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy Score: {}".format(accuracy_score(y_test,y_pred)))
print("Classification_report:\n\n",(classification_report(y_test,y_pred)))
````

````
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus

dot_data = StringIO()

index = ["Pregnancies",	"Glucose" ,	"BloodPressure",	"SkinThickness",	"Insulin",
         "BMI",	"DiabetesPedigreeFunction","Age	"]

export_graphviz(clf, out_file=dot_data, 
                filled=True, rounded=True,
                special_characters=True, feature_names=index, class_names=["0","1"])

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("diabetes.png")
Image(graph.create_png())
````
